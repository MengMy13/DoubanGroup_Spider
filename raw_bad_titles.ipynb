```python
#导入库
import requests
from lxml import etree
import pandas as pd
from fake_useragent import UserAgent
```

```python
url = 'https://www.douban.com/group/699153/discussion?start=0&type=essence'
#小组需要爬取的首页链接
```

```python
headers={
    'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/125.0.0.0 Safari/537.36 Edg/125.0.0.0'
}
#请求头
```

```python
#发送一个请求到目标网址，用给定请求头
res=requests.get(url=url,headers=headers)
#打印返回状态码，检验
print(res.status_code)
```

```python
#使用lxml.etree库中的HTML方法来解析res.text即之前请求返回的网页内容
html=etree.HTML(res.text)
#检验所需要的标题在网页中的地方，通过xpath来获取
lis=html.xpath('//*[@id="content"]/div/div[1]/div[2]/table/tr')
#打印lis列表长度
print(len(lis))
```

```python
def get_first_text(list):
    try:
        return list[0].strip()  # 返回第一个字符串，除去两端的空格
    except:
        return ""
```

```python
#给定网址，并且根据网页间链接规律设置循环，共59页
urls = ['https://www.douban.com/group/699153/discussion?start=0&type=essence'.format(str(i * 25)) for i in range(59)]
count = 1
#创建一个空dataframe对象，有序号和标题列
df=pd.DataFrame(columns=["序号","标题"])

#循环，遍历urls中每个url
for url in urls:
    res = requests.get(url=url, headers=headers)  #发起请求
    html = etree.HTML(res.text)  #将返回的文本加工为可以解析的html
    lis=html.xpath('//*[@id="content"]/div/div[1]/div[2]/table/tr')  #获取tr元素
    
    for li in lis:
        title=get_first_text(li.xpath('./td[1]/a/text()'))  #获得帖子标题
        print(count, title)  # 输出

        df.loc[len(df.index)]=[count,title] #添加到df中

        count += 1 #更新计数（序号）

df.to_excel("travelbadtitle.xlsx",sheet_name="travelbad",na_rep="") #存到excel文件中
```
